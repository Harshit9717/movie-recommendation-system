```python
clean_df['homepage'] = clean_df['homepage'].fillna('')
clean_df['tagline'] = clean_df['tagline'].fillna('')
clean_df['runtime'] = clean_df['runtime'].dropna()
clean_df['release_date'] = clean_df['release_date'].dropna()
clean_df['overview'] = clean_df['overview'].dropna()
clean_df.dropna(inplace=True)
```

```python
def top_popular_movies(df, percentile=0.85, top_n=50):

    C = df['vote_average'].mean()
    M = df['vote_count'].quantile(percentile)

    qualified = df[df['vote_count'] >= M].copy()

    V = qualified['vote_count']
    R = qualified['vote_average']

    qualified['score'] = ((R * V) + (C * M)) / (V + M)

    return qualified.sort_values('score', ascending=False).head(top_n)
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
clean_df['genres']=clean_df['genres'].apply(convert)
clean_df['keywords']=clean_df['keywords'].apply(convert)
```

```python
def convert_cast(text):
    L = []
    count=0
    for i in ast.literal_eval(text):
      if count<3:
        L.append(i['name'])
        count+=1
    return L
```

```python
clean_df['cast']=clean_df['cast'].apply(convert_cast)
```

```python
def convert_crew(text):
    L = []
    for i in ast.literal_eval(text):
      if i['job']=='Director':
        L.append(i['name'])
    return L
```

```python
clean_df['crew']=clean_df['crew'].apply(convert_crew)
```

```python
def collapse(L):
  l1=[]
  for i in L:
    l1.append(i.replace(" ",""))
  return l1
```

```python
clean_df['genres']=clean_df['genres'].apply(collapse)
clean_df['keywords']=clean_df['keywords'].apply(collapse)
clean_df['cast']=clean_df['cast'].apply(collapse)
clean_df['crew']=clean_df['crew'].apply(collapse)
```

```python
clean_df['overview'] = clean_df['overview'].fillna('').apply(lambda x: x.split())
```

```python
clean_df = clean_df.rename(columns={'crew': 'Director'})
```

```python
clean_df['tags']=clean_df['overview']+clean_df['genres']+clean_df['keywords']+clean_df['cast']+clean_df['Director']
```

```python
clean_df['tags']
```

```python
new=clean_df.drop(columns=['overview','genres','keywords','cast','Director','original_language','popularity','release_date','revenue','title_x','vote_average','vote_count','title_y','weighted_rank'])
```

```python
new['tags']=new['tags'].apply(lambda x:" ".join(x))
```

```python
new.head(5)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf=TfidfVectorizer(stop_words='english',max_features=5000)

vector=tfidf.fit_transform(new['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vector)
```

```python
def top_content_recommend(movie):
    index = new[new['original_title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].original_title)
```

```python
top_content_recommend('Gandhi')
```
